```python
import pandas as pd        #library used for processing table or excel,csv datasets....
```

```python
dataset=pd.read_csv("insurance_pred.csv")    # data collection
#dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True) #data preprocessing
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]          #spliting input and output
```

```python
from sklearn.model_selection import GridSearchCV      # it will automatically do train_test_split and model creation steps in traditional method.
from sklearn.ensemble import RandomForestRegressor
params_grid = {"criterion":["squared_error","absolute_error","friedman_mse","poisson"]}
                                            #here we are using dictionary({}) to store hyper tunned parameter.
grid = GridSearchCV(RandomForestRegressor(),params_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
results=grid.cv_results_
#print(result)
table= pd.DataFrame.from_dict(results)  #here we are using dictionary extension because we stored every hyper tunning parameter in dictionary above..
table                 #it will show the every combination of hyper tunned parameter and best model which gridcv find above , in a table here..


# it also used to compare this algorithm with different algorith using rank_test_score which is the best algorithm for this 
#given dataset and problem statement....
```

```python
future_prediction = grid.predict([[19,27.900,0,False,True]])
print("future_prediction:{}".format(future_prediction))          # this is == to r2_score in traditional method.
```

```python
import pickle
filename="finalized_Grid_random_forest.sav"   # saving our best model by a new filename.
pickle.dump(grid,open(filename,"wb"))
```
